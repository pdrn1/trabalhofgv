```python
#Importando as bibliotecas necessrias.
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

#Lendo e guardando as tabelas de dados base.
dados_tratados = pd.read_csv("dados_tratados.csv")

#Mostra a tabela.
dados_tratados

#Lendo e guardando a tabela com a quantidade de provimentos por ano
num_prov = pd.read_csv("numero_provimentos.csv")
num_prov
#Criar grfico de pizza com os dados da tabela de provimentos.
dp = pd.DataFrame(num_prov)
dp["ANO"] = dp["ANO"].astype(int)
soma_prov = dp["Provimento"].sum()
soma_nprov = dp["No provimento"].sum()
soma_provido_parte = dp["Provido em parte"].sum()

# Dados para o grfico de pizza
labels = ["Provimento", "No Provimento", "Provido em parte"]
sizes = [soma_prov, soma_nprov, soma_provido_parte]
colors = ['#66b3ff', '#ff9999', '#A020F0']

# Criando o grfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Provimento vs No Provimento entre 2020 - 2024")
plt.axis("equal")
plt.show()
```

```python
#Mostrar a quantidade de vezes que funo social foi citada em cada ano
dadoAno = pd.read_csv("funcao_social_tratada.csv")
df = pd.DataFrame(dadoAno)
plt.figure(figsize=(10, 6))
plt.plot(df['Ano'], df["Funo social citada"], marker="o", label="Funo social citada")

# Adicionando rtulos e ttulo
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.title("Nmero de vezes que a Funo Social foi Citada por Ano")
plt.legend()

# Exibindo o grfico
plt.grid(True)
plt.show()
```

```python
#Ler e gravar tabela com a quantidade tratada por funo social.
fs = pd.read_csv("funcao_social.csv")

#Mostrar a relao entre a citao de funo social e a provimento da sentena com um grfico de barras empilhadas
fg = pd.DataFrame(fs)
colors = ["#66b3ff", "#ff9999"]
plt.figure(figsize=(10,7))
plt.bar(fg["Tipo"], fg["Quantidade Com FS"], color=colors[0], label=("Com FS")
plt.bar(fg["Tipo"], fg["Quantidade Sem FS"], bottom=fg["Quantidade Com FS"], color=colors[1], label="Sem FS")
#Adicionando rtulos e ttulo
        
plt.ylabel("Quantidade de julgamentos")
plt.title("Resultado")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
#Lendo e salvando a planilha dos desembargadores
dadosDesembarg = pd.read_csv("desembargador.csv")

#Criando o data frame para manipular e criar o grfico
df =pd.DataFrame(dadosDesembarg)
#Crio o grfico

df.set_index("Desembargador", inplace=True)
plt.figure(figsize=(10, 6)) 

#Adicionar rtulos e ttulo
plt.title("Citao de Funo Social por Desembargador")
plt.xlabel("Desembargador")
plt.xticks(rotation=45, ha="right")
plt.bar(df.index, df["Cita Funo Social"], alpha = 1, label="Quantidade de menes a funo social", color="red" ,width=0.5)
plt.plot(df.index, df["Provido"], label = "Quantidade de pedidos providos", color = "blue", linestyle = "--")
plt.plot(df.index, df[" No provido"], label = "Quantidade de pedidos no providos", color = "green", linestyle = ":")
plt.plot(df.index, df[" Provido em parte"], label = "Quantidade de pedidos parcialmente providos", color = "purple", linestyle = "-.")
plt.tight_layout()
plt.grid(True)
plt.legend(loc="best")
plt.show()
```
